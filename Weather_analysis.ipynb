```python
import requests
import json
import random
import matplotlib.pyplot as plt
import pandas as pd
from citipy import citipy
from config import apikey
```

```python
queryurl = 'https://api.openweathermap.org/data/2.5/weather?q={},{}&APPID={}&units={}'
units = 'imperial'
```

```python
lon = [x for x in range(-180, 180, 30)]
lat = [x for x in range(-90, 90, 5)]

queryDict = {
    'cities': [],
    'lat': [],
    'temp': []
}


for lonnum in lon:
    for latnum in lat:
        city = citipy.nearest_city(lonnum, latnum)
        queryCity = city.city_name
        queryCountryCode = city.country_code
        if queryCity not in queryDict['cities']:
            #query weather API and build dict
            weatherResponse = requests.get(queryurl.format(queryCity, queryCountryCode, apikey, units))
            if weatherResponse.json()['cod'] == '404':
                continue
            else:
                print(queryDict)
                queryDict['cities'].append(queryCity)
                queryDict['lat'].append(latnum)
                queryDict['temp'].append(weatherResponse.json()['main']['temp'])
```

```python
weatherData = pd.DataFrame(queryDict)
weatherData.head()
#temperature is in farenheit
```

```python
plt.scatter(weatherData['lat'], weatherData['temp'], marker="o")

plt.title("temperature of cities")
plt.ylabel("Temperature (Farenheit)")
plt.xlabel("Latitude")
plt.grid(True)

plt.show()
```
